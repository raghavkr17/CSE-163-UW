## fill_missing_attack_with_mean

One of the missing values in the dataset is the attack (atk) stat for a Dragonair and a Staravia.
Since attack values tend to be similar within the same Pokemon species, a reasonable way to fill in a missing attack value is to use the mean attack of that species.

I use the pandas groupby and mean functions to calculate the average attack for each Pokemon name. Then, I fill in the missing attack values using that mean. This makes sense because it keeps the value realistic and consistent with the rest of the data.

```python
def fill_missing_attack_with_mean(df):
    """
    Fills missing attack (atk) values using the mean attack
    of Pokemon with the same name.
    """
    mean_attack = df.groupby("name")["atk"].mean()

    df["atk"] = df["atk"].fillna(
        df["name"].map(mean_attack)
    )

    return df
```

```python
data = fill_missing_attack_with_mean(data)
data
```

One of the missing values in the dataset is the attack (atk) stat for a Dragonair and a Staravia.
Since attack values tend to be similar within the same Pokemon species, a reasonable way to fill in a missing attack value is to use the mean attack of that species.

I use the pandas groupby and mean functions to calculate the average attack for each Pokemon name. Then, I fill in the missing attack values using that mean. This makes sense because it keeps the value realistic and consistent with the rest of the data.

```python
def fill_missing_stage_with_min(df):
    """
    Fills missing stage values using the minimum stage
    of Pokemon with the same name.
    """
    min_stage = df.groupby("name")["stage"].min()

    df["stage"] = df["stage"].fillna(
        df["name"].map(min_stage)
    )

    return df
```

```python
data = fill_missing_stage_with_min(data)
data
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load CSV
df = pd.read_csv("nces-ed-attainment.csv")
df.replace("---", pd.NA, inplace=True)
df.set_index(["Year", "Sex", "Min degree"], inplace=True)
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors="coerce")

# Example 1: Heatmap for 'Asian'
heat_data = df.loc[(slice(None), "A", slice(None)), "Asian"].unstack(level="Min degree")
sns.heatmap(heat_data.T, cmap="YlGnBu", annot=True, fmt=".1f")
plt.xlabel("Year")
plt.ylabel("Min degree")
plt.title("Asian educational attainment over time")
plt.show()
```

For the above visualization, answer the following:

### Why did you choose this visualization type?

I wanted a clear way to see differences across Min degrees over time for the Asian population. The heatmap makes trends and gaps easy to spot.



### What is effective about this visualization choice?

It clearly shows which years and degrees have higher or lower percentages. Patterns and missing data are easy to see at a glance.


### What is ineffective about this visualization choice?

It’s hard to compare exact numbers without hovering or looking at the labels. Also, small differences might not stand out visually.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load CSV
df = pd.read_csv("nces-ed-attainment.csv")
df.replace("---", pd.NA, inplace=True)
df.set_index(["Year", "Sex", "Min degree"], inplace=True)
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors="coerce")

# Example 2: Area plot for 'Hispanic'
area_data = df.loc[(slice(None), "A", slice(None)), "Hispanic"].reset_index()
area_pivot = area_data.pivot(index="Year", columns="Min degree", values="Hispanic")
area_pivot.plot(kind="area", stacked=True)
plt.xlabel("Year")
plt.ylabel("Percentage")
plt.title("Hispanic educational attainment over time")
plt.show()
```

For the above visualization, answer the following:

### What type of visualization did you choose?

Area plot.


### Why did you choose this visualization type?

I wanted to show cumulative trends for Hispanic educational attainment over time. The area plot shows how the total percentage builds up across Min degrees.



### What is effective about this visualization choice?

It clearly shows how attainment changes over time and lets you see the contribution of each degree to the total.


### What is ineffective about this visualization choice?

Exact values are harder to read than a bar plot. Overlapping areas can make it difficult to see small changes in lower percentages.

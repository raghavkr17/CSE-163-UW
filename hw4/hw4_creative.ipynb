```python
import networkx as nx
from typing import List


class ProfessionalNetwork(nx.Graph):
    """
    Graph representing a professional networking platform.
    Nodes store job + skills information.
    """

    def add_user(self,
                 name: str,
                 job_title: str,
                 skills: List[str]) -> None:
        """
        Adds a user with attributes.
        """
        self.add_node(
            name,
            job_title=job_title,
            skills=skills
        )

    def connect_users(self,
                      user1: str,
                      user2: str) -> None:
        """
        Connects two professionals.
        """
        self.add_edge(user1, user2)

    def recommend_by_skill(self,
                           user: str) -> List[str]:
        """
        Recommends users with overlapping skills.
        """
        user_skills = set(self.nodes[user]["skills"])
        recs = []

        for person in self.nodes:
            if person == user or self.has_edge(user, person):
                continue

            skills = set(self.nodes[person]["skills"])

            if user_skills & skills:
                recs.append(person)

        return recs

    def get_job_titles(self) -> List[str]:
        """
        Returns all job titles in the network.
        """
        return [
            self.nodes[n]["job_title"]
            for n in self.nodes
        ]
```

```python
net = ProfessionalNetwork()

net.add_user(
    "Raghav",
    "Student Developer",
    ["Python", "SQL", "AI"]
)

net.add_user(
    "Alex",
    "ML Engineer",
    ["Python", "Machine Learning"]
)

net.add_user(
    "Jordan",
    "Product Designer",
    ["Figma", "UX"]
)

net.connect_users("Raghav", "Alex")

print("Skill Recommendations:",
      net.recommend_by_skill("Raghav"))

print("All Job Titles:",
      net.get_job_titles())
```

I chose to extend the Graph class to model a professional networking platform similar to LinkedIn. I thought this was a natural extension of the social network analysis we did in the technical component, but applied in a career setting.

I added node attributes for job_title and skills because those are two of the most important pieces of information when building professional connections. Skills help determine collaboration potential, while job titles show experience level and domain.

For functionality, I implemented methods to add users, connect professionals, recommend connections based on shared skills, and retrieve all job titles in the network. The recommendation function demonstrates how graphs can be used beyond friendships — in this case, for career networking and recruiting.

I imagine this kind of class being used in hiring platforms, alumni networks, or internal company systems to help employees find collaborators based on expertise.
